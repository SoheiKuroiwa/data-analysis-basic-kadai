```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# Step 1: 必要なライブラリのインポートとCSVファイルの読み込み
df = pd.read_csv('sample_data/sample_pandas_6 (1).csv')

# Step 2: カテゴリー列のユニークな出現頻度をカウントし、棒グラフ化する
# 商品名の出現頻度をカウント
product_counts = df['商品名'].value_counts()
print("▼ 商品名ごとの出現頻度")
print(product_counts)

# 棒グラフを作成
plt.figure(figsize=(12, 8)) # グラフのサイズを調整
product_counts.plot(kind='bar')
plt.title('商品名ごとの出現頻度')
plt.xlabel('商品名')
plt.ylabel('出現回数')
plt.xticks(rotation=45, ha='right') # X軸ラベルを斜めにして見やすくする
plt.tight_layout() # レイアウトを自動調整してラベルが見切れないようにする

# Step 3: グラフを表示させる
plt.show()

# Step 4: 基準列ごとに要素をグループにまとめサブ列を集計し、基本統計量を算出する
# 商品番号ごとに注文数の基本統計量を算出
product_stats = df.groupby('商品番号')['注文数'].describe()
print("\n▼ 商品番号ごとの注文数の基本統計量")
print(product_stats)
```
