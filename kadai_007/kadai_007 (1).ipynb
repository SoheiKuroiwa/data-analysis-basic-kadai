```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib # Google Colabなどで日本語を表示するために使用

# --- ファイルの読み込みとマージ ---
# 注文データとカテゴリーデータをそれぞれ読み込む
sales_df = pd.read_csv('sample_data/sample_pandas_6 (1).csv')
category_df = pd.read_csv('sample_data/category (1).csv')

# 「商品番号」をキーとして2つのデータを結合する
# これにより、注文データに「カテゴリー」列が追加される
merged_df = pd.merge(sales_df, category_df, on='商品番号')


# --- Step 2: カテゴリーの出現頻度をカウントし、棒グラフ化 ---
# カテゴリーごとの出現頻度を計算
category_counts = merged_df['カテゴリー'].value_counts()
print("▼ カテゴリーごとの出現頻度")
print(category_counts)

# 棒グラフを作成
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('カテゴリーごとの出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('出現回数')
plt.xticks(rotation=0)
plt.tight_layout()


# --- Step 3: グラフの表示 ---
plt.show()


# --- Step 4: 商品番号ごとの基本統計量を算出 ---
product_stats = merged_df.groupby('商品番号')['注文数'].describe()
print("\n▼ 商品番号ごとの注文数の基本統計量")
print(product_stats)
```
